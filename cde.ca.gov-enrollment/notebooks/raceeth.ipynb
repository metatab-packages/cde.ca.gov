```python
enrollment_07_18_df = pkg.resource('enrollment_07_18').dataframe()
```

```python
col_map = {
    0:'na',
    1:'aian',
    2:'asian',
    3:'pi',
    4:'filipino',
    5:'hisp',
    6:'black',
    7:'nh_white',
    8:'unk', # Not in data dictinoary
    9:'two_more'
}

t = enrollment_07_18_df[['cds_code', 'ethnic','enr_total']]
raceeth = t.groupby(['cds_code', 'ethnic']).sum().unstack()

raceeth.columns = [col_map.get(e,e) for e in raceeth.columns.levels[1]]
raceeth = raceeth.rename({'cds_code':'cdscode'})
```
